## Aplicando Econding nas variáveis
Agora, iremos transformas os valores *object* em valores numéricos binários. Desta forma, o processo de análise dos dados, bem como a criação do modelo de Machine Learnng serão facilitados. 

Para as variáveis que possuem respostas somente "Yes" e "No", vamos apenas substituir por 1 e 0.

```python
for coluna in df.columns:
    print(df[coluna].value_counts())
```

```python
colunas1 = ['Churn', 'customer.Partner', 'customer.Dependents', 'phone.PhoneService', 'account.PaperlessBilling']
for coluna in colunas1:
  df[coluna] = df[coluna].map({'No' : 0, 'Yes' : 1})
```

```python
df
```

A coluna "account.Contract" vamos considerar como valores ordinais, variando entre 0, 1 ou 2 anos:

```python
df['account.Contract'] = df['account.Contract'].map({'Month-to-month' : 0, 'One year' : 1, 'Two year' : 2})
df.head()
```

E no restante das colunas, vamos representar as respostas em formato binário, utilizando a função **get_dummies** do Pandas. Optou-se por usar esta função do Pandas ao invés de utilizar a função **OneHotEncoder** da biblioteca Sklearn, pois gostaríamos de manter o formato DataFrame do dados e trabalhar com os nomes das colunas, de forma que seja possível analisar as variáveis de forma visual e possíveis relações entre elas . Desta forma, poderemos gerar tirar *insights* antes de criar o modelo de ML. 

```python
colunas2 = ['customer.gender', 
            'phone.MultipleLines', 
            'internet.InternetService', 
            'internet.OnlineSecurity', 
            'internet.OnlineBackup', 
            'internet.DeviceProtection', 
            'internet.TechSupport', 
            'internet.StreamingTV', 
            'internet.StreamingMovies',  
            'account.PaymentMethod'
            ]
for coluna in colunas2:
  transforma = pd.get_dummies(df[coluna], prefix = coluna)
  df= pd.concat([df, transforma], axis=1).drop([coluna], axis=1)
```

```python
df.head()
```

```python
df.info()
```

Todas as colunas que representam informações referentes a "No phone service" e "No internet service" são redundantes, pois essa informação já está sendo representada nas colunas "phone.PhoneService" e "internet.InternetService_No", respectivamente. Abaixo temos um exemplo de como as duas colunas são perfeitamente opostas. 

```python
redundancia = [df['phone.PhoneService'], df['phone.MultipleLines_No phone service']]
pd.DataFrame(redundancia)
```

O mesmo ocorre para as colunas com "Yes" e "No" de um mesmo serviço, são perfeitamente opostas e portanto, com informações redundantes. Segue o exemplo:

```python
redundancia2 = [df['internet.OnlineSecurity_Yes'], df['internet.OnlineSecurity_No']]
pd.DataFrame(redundancia)       
```

Portanto, para deixar nosso dados mais leves, todas estas colunas com informações redundantes serão exluídas. 

```python
df = df.drop(['phone.MultipleLines_No phone service',
              'phone.MultipleLines_No',
              'internet.OnlineSecurity_No internet service',
              'internet.OnlineBackup_No internet service',
              'internet.DeviceProtection_No internet service',
              'internet.TechSupport_No internet service',
              'internet.StreamingTV_No internet service',
              'internet.StreamingMovies_No internet service',
              'internet.OnlineSecurity_No',
              'internet.OnlineBackup_No',
              'internet.DeviceProtection_No',
              'internet.TechSupport_No',
              'internet.StreamingTV_No',
              'internet.StreamingMovies_No'
              ], axis=1)
```

```python
df.info()
```

Desta forma, foi possível reduzir nossos dados de 40 para 26 colunas, sem perda de informação.
