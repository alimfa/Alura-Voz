# Análise Exploratória e Visualização do Dados

## Análise Perfil dos Clientes

```python
#Plota gráficos com perfil dos clientes
plt.figure(figsize = ((15, 10)))

#Gráfico 1 - Genero
x = ['Feminino', 'Masculino']
y = [data['Gen_Feminino'].sum(), data['Gen_Masculino'].sum()]
plt.subplot(2, 2, 1)
f1 = plt.bar(x,y, color =['#40b7ad', '#8bdab2'])
plt.bar_label(container= f1, label_type='edge')
plt.ylim(top=7100)
plt.title("Gênero", fontsize = 16)

#Gráfico 2 - Idosos
x = ['Não Idoso', 'Idoso']
y = [data.Idoso[data.Idoso == 0].count(), data.Idoso[data.Idoso == 1].count()]
plt.subplot(2, 2, 2)
f2 = plt.bar(x,y, color = ['#37659e', '#348fa7'])
plt.bar_label(container= f2, label_type='edge')
plt.ylim(top=7100)
plt.title("Idosos", fontsize = 16)

#Gráfico 3 - Parceiros
x = ['Sem Parceiro', 'Com Parceiro']
y = [data.Parceiro[data.Parceiro == 0].count(), data.Parceiro[data.Parceiro == 1].count()]
plt.subplot(2, 2, 3)
f3 = plt.bar(x,y, color = ['#37659e', '#348fa7'])
plt.bar_label(container= f3, label_type='edge')
plt.ylim(top=7100)
plt.title("Parceiros", fontsize = 16)

#Gráfico 4 - Dependentes
x = ['Sem Dependentes', 'Com Dependentes']
y = [data.Dependentes[data.Dependentes == 0].count(), data.Dependentes[data.Dependentes == 1].count()]
plt.subplot(2, 2, 4)
f4 = plt.bar(x,y, color = ['#37659e', '#348fa7'])
plt.bar_label(container= f4, label_type='edge')
plt.ylim(top=7100)
plt.title("Dependentes", fontsize = 16)

plt.suptitle("PERFIL DOS CLIENTES", fontsize = 20)

plt.subplots_adjust(left=0.125, bottom=0.1, right=0.9, top=0.9, wspace=0.2, hspace=0.35)
plt.show()
```

A partir dos gráficos apresentados podemos entender que nossa amostra está bastante equilibrada entre clientes do gênero masculino e feminino, assim como em relação à possuírem ou não parceiros.

Além disso, percebe-se que entre todos os clientes analisados, 16% são idosos e 30% possuem dependentes.

```python
#Cria variável com o número de clientes que possuem combo telefone + internet
combo = 0
for i, row in data.iterrows():
  if (row['Servico_Telefone'] == 1) and (row['Sem_Servico_Internet'] == 0):
    combo = combo + 1
combo

#Plota Gráfico dos Tipos de Serviços
plt.figure(figsize = ((20, 15)))
x = np.arange(3) 
y1 = [data.Servico_Telefone[data.Servico_Telefone == 0].count(), data.Sem_Servico_Internet[data.Sem_Servico_Internet == 1].count(), 0]
y2 = [data.Servico_Telefone[data.Servico_Telefone == 1].count(), data.Sem_Servico_Internet[data.Sem_Servico_Internet == 0].count(), combo]
plt.subplot(2, 2, 1)
plt.bar(x, y1, color = '#413d7b', bottom = y2) 
plt.bar(x, y2, color = '#37659e') 
plt.xticks(x, ['Telefone', 'Internet', 'Telefone e Internet']) 
plt.title("Tipos de Serviços", fontsize = 16)
plt.legend(["Não possui", "Possui"])
plt.show() 
```

Percebe-se que a maioria dos clientes possui combo com serviço de internet e telefone juntos, inclusive, mais de 90% dos clientes possui serviço de telefone.

## Análise taxa de Evasão dos clientes

Primeiramente, é necessário entender a porcentagem de clientes que cancelaram seus planos, e posteriormente, observar a relação entre nossa variável **target** e as demais, a fim de entender os possíveis motivos que levaram ao cancelamento e verificar alguns padrões.

```python
#Plota gráfico da taxa de evasão
plt.figure(figsize = ((6, 6)))

plt.pie([data.Evasao[data.Evasao == 0].count(), data.Evasao[data.Evasao == 1].count()], 
        labels=['Continuou', 'Cancelou'], 
        colors=['#348fa7', '#8bdab2'],
        autopct='%.0f%%', 
        explode = (0, 0.05))

plt.suptitle("EVASÃO DE CLIENTES", fontsize = 15)
plt.show()
```

Dos 7043 clientes analisados, 27%, ou mais precisamente, 1869 cancelaram seus planos. 

Para ter uma visão geral e entender melhor todas as relações entre as diferentes variáveis, vamos plotar uma matriz correlação:


```python
#Cria novo DataFrame, excluindo primeira coluna com ID dos clientes
data3 = data.drop('ID', axis = 1)
data3
```

```python
#Plota matriz correlação
plt.figure(figsize = ((15, 15)))
mask = np.triu(np.ones_like(data3.corr())) 
dataplot = sns.heatmap(data3.corr(), cmap="YlGnBu", annot=True, mask=mask) 
plt.show() 
```

- Fazendo a correlação entre todas as variáveis e a Evasão, o que mais chama atenção em relaçãoé: a Fibra Otica, Meses de Contrato, Cheque eletrônico e tipo de contrato. 

- Foram percebidas outras correlações não óbivas entre as variáveis: 

 - internet fibra ótica x fatura digital
 - internet fibra ótica x valor mensal
 - Fibra Otica x múltiplas linhas


Relacionando a variável **target** com o perfil dos clientes, podemos perceber alguns pontos de atenção:

```python
#Plota gráficos que mostram a relação entre o perfil dos clientes e a taxa de evasão
plt.figure(figsize = ((15, 10)))

#Gráfico 1 - Genero x Evasão
plt.subplot(2, 2, 1)
a1 = sns.countplot(x = 'Gen_Masculino', 
              hue = 'Evasao', 
              data = data, 
              palette = ['#348fa7','#8bdab2'], 
              dodge =  False)
a1.set_xticklabels(['Feminino', 'Masculino'])
a1.set_xlabel(None)
a1.set_ylabel(None)
plt.legend(labels=['Continuou', 'Cancelou'])
plt.title('Gênero', fontsize = 16)

#Gráfico 2 - idosos x Evasão
plt.subplot(2, 2, 2)
a2 = sns.countplot(x = 'Idoso', 
              hue = 'Evasao', 
              data = data, 
              palette = ['#348fa7','#8bdab2'], 
              dodge =  False)
a2.set_xticklabels(['Não Idoso', 'Idoso'])
a2.set_xlabel(None)
a2.set_ylabel(None)
plt.legend(labels=['Continuou', 'Cancelou'])
plt.title('Idosos', fontsize = 16)

#Gráfico 3 - Parceiros x Evasão
plt.subplot(2, 2, 3)
a3 = sns.countplot(x = 'Parceiro', 
              hue = 'Evasao', 
              data = data, 
              palette = ['#348fa7','#8bdab2'], 
              dodge =  False)
a3.set_xticklabels(['Sem Parceiro', 'Com Parceiro'])
a3.set_xlabel(None)
a3.set_ylabel(None)
plt.legend(labels=['Continuou', 'Cancelou'])
plt.title('Parceiro', fontsize = 16)

#Gráfico 4 - Dependentes x Evasão
plt.subplot(2, 2, 4)
a4 = sns.countplot(x = 'Dependentes', 
              hue = 'Evasao', 
              data = data, 
              palette = ['#348fa7','#8bdab2'], 
              dodge =  False)
a4.set_xticklabels(['Sem Dependentes', 'Com Dependentes'])
a4.set_xlabel(None)
a4.set_ylabel(None)
plt.legend(labels=['Continuou', 'Cancelou'])
plt.title('Dependentes', fontsize = 16)


plt.suptitle("PERFIL DOS CLIENTES x EVASÃO", fontsize = 20)
plt.subplots_adjust(left=0.125, bottom=0.1, right=0.9, top=0.9, wspace=0.2, hspace=0.35)
plt.show()
```

- O gênero não influencia na "Churn Rate"
